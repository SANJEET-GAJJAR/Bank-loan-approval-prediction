```python
#seaborn uses mathlib underneath so import mathlib first.
#we use seaborne only when graphs overlap unless then we dont use it
# pandas for managing datasets
import pandas as pd
# matplotlib for additional customization
from matplotlib import pyplot as plt
#%matplotlib inline
# seaborn for plotting and styling
import seaborn as sns
```

```python
# read dataset
df = pd.read_csv('C:/Users/prate/Downloads/Dataset1.csv', index_col=0)
```

```python
# display first 5 observations
df.head()
```

```python
# scatter plot
plt.figure()
sns.lmplot(data=df, x='ApplicantIncome', y='LoanAmount')
plt.show()
```

```python
# now extras
plt.figure()
sns.lmplot(x='ApplicantIncome', y='LoanAmount', data=df,
           fit_reg=False, # No regression line jo upar dikh rahi thi
           hue='Property_Area')# hue gives colour to the plotted points on abu criteria here it is Class column from excel file
plt.show()
```

```python
# now extras
plt.figure()
sns.lmplot(x='ApplicantIncome', y='LoanAmount', data=df,
           fit_reg=False, # No regression line jo upar dikh rahi thi
           hue='Property_Area')# hue gives colour to the plotted points on abu criteria here it is Class column from excel file
# tweak using matplotlib
plt.ylim(-10, None)#-10 says that there is a shift of origin by -10,-10
plt.xlim(-10000, None)
#plt.title('Graph Title')
#plt.ylabel('Y axis')
#plt.xlabel('X axis')
# good practice
plt.show()
```

```python
# boxplot
plt.figure()
sns.boxplot(data=df)
plt.show()
#boxplot will give 5 point summary. it generates box plot for every numeric column
```

```python
# temp df with select cols. and shows the rest columns
tdf = df.drop(['LoanAmount', 'Credit_History','Loan_Amount_Term'], axis=1)
tdf.head()
# new boxplot using stats_df
plt.figure()
sns.boxplot(data=tdf)
plt.show()
```

```python
# histogram
plt.figure()
sns.distplot(df.ApplicantIncome, bins=7, color='r')#specify the number of bins and not size
#sns.distplot(df.ApplicantIncome, kde=False, bins=6, color='g')#no line
sns.distplot(df.ApplicantIncome, hist=False, kde=True, bins=7, color='r', vertical=False)#only line
# #vrtical canges the axis and makes it horizontal if it is True
plt.show()
```

```python
# frequency plot 
plt.figure()
sns.countplot(x='Property_Area', data=df) 
# Rotate x-labels
plt.xticks(rotation=-45)
plt.show()
#works with catrgoric data while histogram works with numeric data.
```

```python
# basic colors
# https://matplotlib.org/2.0.2/api/colors_api.html
# named colors
# https://matplotlib.org/3.1.0/gallery/color/named_colors.html

# create palette
type_palette = ['#78C850',  # Grass
                '#F08030',  # Fire
                '#6890F0',  # Water
                '#A8B820',  # Bug
                '#A8A878',  # Normal
                '#A040A0',  # Poison
                '#F8D030',  # Electric
                '#E0C068',  # Ground
                '#EE99AC',  # Fairy
                '#C03028',  # Fighting
                '#F85888',  # Psychic
                '#B8A038',  # Rock
                '#705898',  # Ghost
                '#98D8D8',  # Ice
                '#7038F8',  # Dragon
                ]
#if we omit one colur it will start asssigning colour again from first. This can ve dangerous
```

```python
# frequency plot 
plt.figure()
sns.countplot(x='Property_Area', data=df, palette=type_palette)
# Rotate x-labels
plt.xticks(rotation=45)#rotates the names so that they dont overlap
plt.show()
```

```python
# heatmap
# temp df with select cols
tdf = df.drop(['Gender','Married','Loan_Amount_Term','Credit_History'], axis=1)
tdf.head()
```

```python
# 
corr = tdf.corr()
print(corr)
# calculate correlations
plt.figure()
# plot heatmap
sns.heatmap(corr)#gives colour on bais of values. see the scale besides it.
plt.show()
```
